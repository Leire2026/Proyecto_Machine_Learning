### Justificación   

Después de utilizar una Regresión Lineal como modelo base, resulta interesante probar un Árbol de Decisión para regresión porque permite capturar relaciones no lineales entre las variables explicativas y la satisfacción con la vida.

A diferencia de la regresión lineal, que asume una relación lineal entre las variables predictoras y la variable objetivo, los árboles de decisión dividen los datos en grupos homogéneos mediante reglas de decisión sucesivas. Esto permite identificar patrones más complejos y posibles interacciones entre variables.

Además, una de sus principales ventajas es la interpretabilidad, ya que las reglas generadas por el árbol pueden visualizarse y analizarse fácilmente. Por ello, este modelo constituye una alternativa interesante para evaluar si existen relaciones no lineales que mejoren la capacidad predictiva respecto a la Regresión Lineal.


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
df_modelo = pd.read_csv("../1_Data/processed.csv", sep = ",")

df_modelo
```

```python
df_modelo.info()
```

```python
df_modelo.describe(include='all')
```

```python
# Separo X e y
# Variables predictoras
X = df_modelo.drop(columns=['Target_Satisfacc_Vida'])

# Variable target
y = df_modelo['Target_Satisfacc_Vida']

print("Dimensiones de X:", X.shape)
print("Dimensiones de y:", y.shape)
```

```python
# Separo en train y test
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=10
)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)

print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

Con el objetivo de homogeneizar la escala de las variables predictoras, aplico una estandarización mediante StandardScaler. Esta transformación centra cada variable en media cero y desviación típica uno. El escalador se ajusta exclusivamente sobre el conjunto de entrenamiento y posteriormente se aplica al conjunto de prueba.

```python
# Escalado

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```
